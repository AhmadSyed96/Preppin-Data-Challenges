```python
df = None
df_menu = None
for x in range(0,df_elongated.shape[1] -2, 3):
    df = df_elongated.iloc[:,x:x+3].copy()
    df = df.assign(Category=df.columns[0])
    df.columns = ['Subcategory', 'Price', 'ID', 'Category']
    df_menu = pd.concat([df_menu, df])
df_menu.dropna(inplace=True)
df_menu.reset_index(drop=True, inplace=True)
df_menu
```

```python
df_orders = pd.read_excel('Menu and Orders.xlsx', sheet_name='Order')
df_orders['Order'] = df_orders['Order'].astype('str').str.split('-')
df_orders = df_orders.explode('Order')
df_orders['Order'] = df_orders['Order'].astype('int')
df_orders.reset_index(drop=True, inplace=True)
df_orders
```

```python
df_output_foundation = pd.merge(df_orders, df_menu, left_on='Order', right_on='ID', how='left')
df_output_foundation['Date'] = pd.to_datetime(df_output_foundation['Order Date'], format='%Y-%m-%d')
df_output_foundation['Day'] = df_output_foundation['Date'].dt.day_name()
df_output_foundation
```

```python
df_max_orders_by_customer = df_output_foundation.groupby('Customer Name', as_index=False)['Subcategory'].count()
df_max_orders_by_customer.loc[df_max_orders_by_customer['Subcategory'] == df_max_orders_by_customer['Subcategory'].max()]
```

```python
df_tot_per_DOWeek = df_output_foundation.groupby('Day')['Price'].sum()
df_tot_per_DOWeek
```
