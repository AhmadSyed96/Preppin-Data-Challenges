```python
import pandas as pd
from pandasgui import show
import numpy as np
```

```python
# load products table
df_products = pd.read_excel('Shopping List and Ingredients.xlsx')
df_products
```

```python
# load animal ingredients table
df_non_veg_ingr = pd.read_excel('Shopping List and Ingredients.xlsx', sheet_name='Keywords')
df_non_veg_ingr
```

```python
# edit the animal table 
df_non_veg_ingr = df_non_veg_ingr.transpose().rename(columns={0:'Animal Ingredient'}).reset_index(drop=True).apply(lambda x: x.str.split(', ')).explode('Animal Ingredient').reset_index(drop=True)
df_non_veg_ingr['Animal Ingredient'] = np.where(df_non_veg_ingr['Animal Ingredient'].str.isnumeric(), 'E'+df_non_veg_ingr['Animal Ingredient'], df_non_veg_ingr['Animal Ingredient'])
df_non_veg_ingr
```

```python
# crossjoin
df_non_veg_ingr['dummy'] = 1
df_products['dummy'] = 1

df_output = pd.merge(df_non_veg_ingr, df_products, how='left', on='dummy')
df_output
```

```python
df_output['non-veg?'] = df_output.apply(lambda x: x['Animal Ingredient'] in x['Ingredients/Allergens'], axis=1)
df_output
```

```python
df_output_veg = df_output.loc[~df_output['non-veg?']]
df_output_veg
```

```python
df_output_veg = df_output.loc[df_output['non-veg?']]
df_output_veg = df_output_veg.groupby(['Product', 'Description', 'Ingredients/Allergens']).agg({'Animal Ingredient':lambda x: ', '.join(x.tolist())})
df_output_veg
```
