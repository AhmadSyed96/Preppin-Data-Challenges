```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Joined Dataset.csv').drop_duplicates(subset = ['Contact Email','Contact Name','Client','Training'])
df
```

```python
df_newest_manager = df.sort_values(by = ['From Date']).groupby('Client')[['Client ID','Account Manager', 'From Date', 
                           ]].last().reset_index()
df_newest_manager
```

```python
df_output = df.merge(df_newest_manager,how='left',on=['Client'])
df_output.drop(['Client ID_x','Account Manager_x','From Date_x'], axis=1, inplace=True)
df_output.rename(columns={'Account Manager_y':'Account Manager','Client ID_y':'Client ID','From Date_y':'From Date'}, inplace=True)
df_output
```
