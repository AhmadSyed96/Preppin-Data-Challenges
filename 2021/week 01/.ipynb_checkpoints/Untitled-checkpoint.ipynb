```python
#import
import pandas as pd
import numpy as np
from timeit import default_timer as timer
```

```python
df_purchaser_info = pd.read_excel('PD 2021 Wk 1 Input.xlsx')
df_purchaser_info
```

```python
# split the Store - Bike column
df_purchaser_info[['Store', 'Bike']] = df_purchaser_info['Store - Bike'].str.split(' - ', expand=True)
df_purchaser_info.drop(columns=['Store - Bike'], inplace=True)
df_purchaser_info
```

### these are the following choices to fix the bike column





```python
# fix the bike column
start = timer()
bikefix_pt = {'Mountain':'^M.*', 'Gravel':'^G.*', 'Road':'^R.*'}
for correct, pattern in bikefix_pt.items():
    df_purchaser_info['Bike'] = df_purchaser_info['Bike'].replace(to_replace = pattern, value = correct, regex = True)
end = timer()
print("The time taken is ", end - start) 
```

```python
# Create a 'Quarter' and 'Day of Month' fields
df_purchaser_info['Date'] = pd.to_datetime(df_purchaser_info['Date'])
df_purchaser_info['Quarter'] = df_purchaser_info['Date'].dt.quarter 
df_purchaser_info['Day of Month'] = df_purchaser_info['Date'].dt.day 
df_purchaser_info
```

```python
# Remove the first 10 orders
df_purchaser_info = df_purchaser_info.loc[(df_purchaser_info['Order ID'] >= 11)]
```
