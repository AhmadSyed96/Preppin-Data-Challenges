```python
import pandas as pd
```

# Steps:
### 1. Compile the tables
### 2. pivot the table and add the needed columns
### 3. get output one via groupby 
### 4. get output two by grouping again

### 1. Compile the tables

```python
with pd.ExcelFile('PD 2021 Wk 3 Input.xlsx') as xl:
    df_uncleaned = pd.concat([pd.read_excel(xl, s).assign(Store=s) for s in xl.sheet_names])
df_uncleaned
```

### 2. pivot the table and add the needed columns

```python
# pivot the table and split
df_cleaned = df_uncleaned.melt(id_vars=['Date', 'Store'], value_vars=['New - Saddles', 'New - Mudguards', 'New - Wheels', 'New - Bags', 'Existing - Saddles', 'Existing - Mudguards', 'Existing - Wheels', 'Existing - Bags'], value_name='Sales Amount', var_name='Type')
df_cleaned[['Customer Type', 'Product']] = df_cleaned['Type'].str.split(' - ', expand=True)
df_cleaned.drop(columns='Type', inplace=True)

# get the quarter
df_cleaned['Quarter'] = df_cleaned['Date'].dt.quarter

df_cleaned
```

### 3. get output one via groupby 

```python
# groupby by product & quarter
df_output_1 = df_cleaned.groupby(['Product', 'Quarter'], as_index=False).agg(Products_Sold=('Sales Amount', 'sum'))
df_output_1.rename(columns=lambda x: x.replace('_', ' '), inplace=True)
df_output_1                                                                         
```

### 4. get output two by grouping again

```python
# group by Store, ustomer type, and Product
df_output_2 = df_cleaned.groupby(['Store', 'Customer Type', 'Product'], as_index=False).agg(Products_Sold=('Sales Amount', 'sum'))
df_output_2.rename(columns=lambda x: x.replace('_', ' '), inplace=True)
df_output_2                                                               
```
