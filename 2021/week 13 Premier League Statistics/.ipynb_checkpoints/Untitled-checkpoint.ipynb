This week we want to create a data set that allows us to analyse 'Open Play Goals' scored. We will rank the players overall and by their position. 

1. Remove all goalkeepers from the data set
2. Remove all records where appearances = 0
3. In this challenge we are interested in the goals scored from open play
    * Create a new “Open Play Goals” field (the goals scored from open play is the number of goals scored that weren’t penalties or freekicks)
    * Note some players will have scored free kicks or penalties with their left or right foot
    * Be careful how Prep handles null fields! (have a look at those penalty and free kick fields) 
    * Rename the original Goals scored field to Total Goals Scored
4. Calculate the totals for each of the key metrics across the whole time period for each player, (be careful not to lose their position)
5. Create an open play goals per appearance field across the whole time period
6. Rank the players for the amount of open play goals scored across the whole time period, we are only interested in the top 20 (including those that are tied for position) – Output 1
7. Rank the players for the amount of open play goals scored across the whole time period by position, we are only interested in the top 20 (including those that are tied for position) – Output 2
8. Output the data – in your solution on twitter / the forums, state the name of the player who was the only non-forward to make it into the overall top 20 for open play goals scored

```python
import pandas as pd
from os import listdir
from pandasgui import show
```

```python
df = pd.concat([pd.read_csv(r'.\inputs\\' + f) for f in listdir(r'.\inputs')]).query('Position != "Goalkeeper" and Appearances > 1')
df
```

```python
df_output_1 = df.groupby(['Name', 'Position'], as_index=False).agg({'Appearances':'sum', 'Goals':'sum', 'Penalties scored':'sum', 'Freekicks scored':'sum', 'Goals with right foot':'sum', 'Goals with left foot':'sum', 'Headed goals':'sum'})
df_output_1['Open Play Goals'] = df_output_1['Goals'] - df_output_1['Penalties scored'] - df_output_1['Freekicks scored']
df_output_1['Rank'] = df_output_1.groupby('Position')['Goals'].rank(ascending=False)
# df_output_1.sort_values('Goals', ascending=False)
show(df_output_1)
```
