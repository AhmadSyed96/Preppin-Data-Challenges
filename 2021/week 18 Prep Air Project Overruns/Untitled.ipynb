```python
import pandas as pd
from timeit import default_timer as timer
```

```python
df= pd.read_excel('PD 2021 Wk 18 Input.xlsx')
df
```

```python
df['Completion Date'] = df['Scheduled Date'] + pd.to_timedelta(df['Completed In Days from Scheduled Date'],'d')
df['Completed Weekday'] = df['Completion Date'].dt.day_name()
```

```python
index=['Owner', 'Project', 'Sub-project']
df_pivoted = df.pivot(index=index, columns='Task', values='Completion Date').reset_index().rename_axis(None, axis=1)
df_pivoted['Scope to Build Time'] = df_pivoted['Build'] - df_pivoted['Scope']
df_pivoted['Build to Delivery Time'] = df_pivoted['Deliver'] - df_pivoted['Build']
df_pivoted.drop(columns=['Build', 'Scope', 'Deliver'], inplace=True)
df_pivoted
```

```python
df_output = pd.merge(df, df_pivoted, on=index)
df_output
```
