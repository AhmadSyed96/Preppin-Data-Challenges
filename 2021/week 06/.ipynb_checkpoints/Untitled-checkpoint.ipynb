```python
import pandas as pd
from pandasgui import show
```

# Steps:
### 1. calculate rankings
### 2. groupby and aggregate
### 3. melt and pivot
### 4. calculate the difference between the two tours

```python
df = pd.read_excel('PGALPGAMoney2019.xlsx').rename(columns=str.title)
df
```

```python
# get rankings
df['Tour Rank'] = df.groupby('Tour')['Money'].rank()
df['Ovr Rank'] = df['Money'].rank()
df['Diff in Ranking'] = df['Ovr Rank'] - df['Tour Rank']
df
```

```python
# grouby
df_output = df.groupby('Tour', as_index=False).agg(Avg_Diff_in_Ranking=('Diff in Ranking','mean'), Avg_Money_per_Event=('Money','mean'), Num_of_Players=('Player Name','count'), Num_of_Events=('Events','sum'), Tot_Prize_Money=('Money','sum'))
df_output.rename(columns=lambda x:x.replace('_', ' '), inplace=True)
df_output
```

```python
# melt
df_output = df_output.melt(id_vars='Tour', var_name='Measure')
df_output
```

```python
# pivot
df_output = df_output.pivot(index=['Measure'], columns='Tour', values='value').reset_index().rename_axis(None, axis=1)
df_output
```

```python
df_output['Difference'] = df_output['LPGA'] - df_output['PGA']
df_output
```
